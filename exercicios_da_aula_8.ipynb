```python
## exercicio 3

import matplotlib.pyplot as plt
import math
""" pendulo simples:
teta_0 = angulo inicial
teta = angulo
l = comprimento 
t = tempo
g =gravidade
"""

g = 9.8
l = 1
periodo_pendulo = 2*math.pi*math.sqrt(l/g)

def angulo_pendulo(teta_0,t) :
    teta = teta_0*math.cos(math.sqrt(g/l)*t)
    return teta
  
def x_pendulo(teta_0,t) :
    teta = angulo_pendulo(teta_0,t)
    x = l*math.sin(teta)
    return x

def y_pendulo(teta_0,t) :
    teta = angulo_pendulo(teta_0,t)
    y = - l*math.cos(teta)
    return y

tempo = 0
lista_tempo=[]
lista_x=[]
lista_y=[]


while tempo <= periodo_pendulo:
    tempo = tempo + periodo_pendulo/100
    lista_tempo.append(tempo)
    
    x = x_pendulo(math.pi/2,tempo)
    lista_x.append(x)
    
    y = y_pendulo(math.pi/2, tempo)
    lista_y.append(y)



plt.plot(lista_x,lista_y)
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.show()

plt.plot(lista_tempo,lista_x)
plt.xlabel("tempo")
plt.ylabel("eixo x")
plt.show()

plt.plot(lista_tempo,lista_y)
plt.xlabel("tempo")
plt.ylabel("eixo y")
plt.show()


```
